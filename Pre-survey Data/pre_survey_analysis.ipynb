#### Reading in the raw data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 

full_data = pd.read_csv('CounterStereotype_Pre-survey_Raw.csv', skiprows = 0, delimiter = ",")
print("original survey data:")
display(full_data)

#removing the first 2 rows and the 1st and 4th column, which only show the question names, metadata, timestamps and consent responses
raw_data = full_data[["Q1:Lawyer","Q2:Lawyer","Q1:Politician","Q2:Politician","Q3","id"]]
raw_data = raw_data.iloc[2:]
raw_data = raw_data.reset_index(drop=True) #resetting the indexes
print("survey data with unnessesary rows and columns removed:")
display(raw_data)

#removing any duplicate ids and only keeping the first response for each unique id
print("survey data with duplicate IDs removed:")
raw_data = raw_data.drop_duplicates(subset='id').reset_index(drop=True)

display(raw_data)
```

Now that the data has been imported and cleaned, the most common traits can be determined.

```python
import matplotlib.pyplot as plt
import seaborn as sns

from collections import Counter

# Function to plot the top N most common words for a given column
def plot_top_words(series, column_name, top_n=20):
    #splitting and count the words in each column
    words = series.str.cat(sep=', ').split(', ')
    #create an array of tuples to store the top n words along with their cooresponding counts
    word_counts = Counter(words).most_common(top_n)
    
    # separating words and their counts
    words, counts = zip(*word_counts)
    
    #plotting the data
    plt.figure(figsize=(4, 4))
    sns.barplot(x=list(counts), y=list(words), palette="Spectral")
    plt.title(f'Top {top_n} Most Common Words in {column_name}')
    plt.xlabel('Frequency')
    plt.ylabel('Words')
    plt.show()

#running the funciton on each row of the dataframe
for column in raw_data.columns:
    if raw_data[column].dtype == object:
        plot_top_words(raw_data[column], column, top_n=20)
```

In order to get a better idea of the freqencies of certain traits, as well as to identify any significant errors, a function is needed to display the traits and their frequencies.

```python
#Function to print the top n words
def print_words(series, column_name, top_n=20):
    words = [word for word in series.str.cat(sep=', ').split(', ')]
    word_counts = Counter(words).most_common(top_n)

    print(column_name)
    print(word_counts)
    print("\n")

#running it on all the data
for column in raw_data.columns:
    print_words(raw_data[column], column, top_n=10000)
```

### Further Cleaning and Analysis
There are a few glaring issues in the preliminary results above. Many of the trait frequencies are being under-reported for a variety of reasons. 

First of all, the frequency is case-sensitive. This means that the frequencies of certain words are being under-reported because the capitalized and non-capitalized versions are counted as two separate words. For example, the trait `tired` had a frequency of 52, and the trait `Tired` had a frequency of 12 in the `Q3` column. Ideally, these frequencies should be lumped together. To fix this, **all capitalization from the responses must be removed.**

Secondly, certain traits are comprised of multiple words, and some respondents chose to write those words with a hyphen between them, while other respondents simply used a space. For example, the trait `hard-working` had a frequency of 5, while the trait `hard working` had a frequency of 13 in the `Q1:Lawyer` column. Ideally, these frequencies should be lumped together, so **all hyphens from the responses should be removed and replaced with spaces**.

Additionally, some of the respondents responded with full sentences rather than just traits. This means words such as `they`, `are`,`to` and `that`,`do`,`will`, should be **removed from the data frame, along with periods.** **Intensifiers such as `very` should be removed as well**, and the word **`and` should be replaced by a comma** so that traits separated by an `and` rather than by a comma will not be counted as a singular trait. 

Due to the way some of the responses are formatted (some respondents seemed to have left erroneous spaces), the most common trait for certain questions such as `Q2:Politician` is shown to be `''` (a blank space). These **blank spaces should be filtered out before counting the frequencies** rather inside the data frame to ensure the different cases where this occurred in the responses do not need to be individually addressed.

At this point, since all duplicate responses have been removed, **the `id` column may also be removed** from the response data frame. 

```python
#removing the id column
cleaned_data = raw_data[["Q1:Lawyer","Q2:Lawyer","Q1:Politician","Q2:Politician","Q3"]]

#lamda funciton to remove the capitalization from all the responses
cleaned_data = cleaned_data.applymap(lambda x: x.lower() if isinstance(x, str) else x)

words_to_remove = [' they','they ',' are ',' to ',' will ',' that',' do',' very']

#this lambda function will replace all the words in the words_to_remove list with '' (removes them)
for word in words_to_remove:
    cleaned_data = cleaned_data.applymap(lambda x:x.replace(word,'') if isinstance(x,str) else x)

#remove all of the uunnecessary punctuation
cleaned_data = cleaned_data.applymap(lambda x: x.replace('-', ' ').replace(' and',',').replace('&',',') if isinstance(x, str) else x)
cleaned_data
```

```python
def plot_top_words2(series, column_name, top_n=20):
    # Split and count the words and remove empty spaces
    words = [word for word in series.str.cat(sep=', ').split(', ') if word != '']
    words = [letter.replace(" ", "") for letter in words]
    word_counts = Counter(words).most_common(top_n)

    words, counts = zip(*word_counts)
    
    # Plot the data and print out the words and frequencies
    plt.figure(figsize=(4, 4))
    print(word_counts)
    sns.barplot(x=list(counts), y=list(words), palette="Spectral")
    plt.title(f'Top {top_n} Most Common Words in {column_name}')
    plt.xlabel('Frequency')
    plt.ylabel('Words')
    plt.show()

for column in cleaned_data.columns:
    if cleaned_data[column].dtype == object:
        plot_top_words2(cleaned_data[column], column, top_n=20)
```

### Results and Future Analysis

There is certainly more that can be done to clean the data further. Certain responses had traits that were variants (inflections) of other traits. For example, the trait `like to argue` (shown above as `likeargue`) had a frequency of 9, and `argumentative` had a frequency of 40 for the `Q1:Lawyer` column. These two traits are clearly variations of the same root word, however it is impossible to group them together using simple code. Instead, this can be done by using a **natural language processing model (NLP) to lemmatize all of the responses into the adjective forms of each word**. This can be done in Python using the open-source Natural Language Toolkit library, however, this is **beyond the scope of this preliminary study**, and would most likely not yield substantially different results.

In a similar vein, a portion of respondents did not use commas to separate their responses, so their entire response was marked as a singular trait rather than the individual traits that they had listed. This was a problem with such a small number of responses that once again, it would most likely not affect the overall frequencies much, especially given the difficulty of cleaning these types of responses algorithmically.

One thing that may significantly impact the frequencies is the presence of synonyms in the data. For example, in the `Q2 Politician` column, `honest` has a frequency of 34, and `truthful` has a frequency of 10. `honest` and `truthful` are synonyms, so one could argue that **synonyms like these should be counted together**. However, there are multiple problems with synonym grouping. 

Most synonyms do not have identical meanings in English. As defined in computational linguistics $ ^{1}$, two words can have varying degrees of synonymy. Synonyms often vary in terms of intensity, like the words `big` and `enormous`, are only synonyms in certain contexts, or only possess somewhat related meanings. **Thus, most words that are considered synonyms by the layperson are only partial synonyms. **

Consider the following traits listed in the `Q2 Politician` column: `honest`, `truthful`, and `fair`. `honest` is a synonym of `truthful`, and it is also a synonym of `fair`. However, `truthful` and `fair` are not synonyms. This is because not all synonym relationships are symmetrical or transitive, as is discussed in the paper above. In fact, there are likely no true synonyms (words with the EXACT meaning in all contexts, and whose other true synonyms are all synonyms of one another) in the English language$ ^{2}$ . Thus, it is **not possible to group synonyms together in a way that is objective.**


Bearing these limitations in mind, **the results of this pre-survey are as follows.** 

The 10 most common perceived **stereotypes for the social group "Lawyer"** among the population of Michigan State Undergraduates are `smart, rich, argumentative, hardworking, rude, greedy, mean, arrogant, stuckup,` and `aggressive`. 

The 10 most common perceived **counter-stereotypes for the social group "Lawyer"** among the population of Michigan State Undergraduates are `lazy, dumb, quiet, shy, empathy, kind, poor, nice, caring,` and `easygoing`.

The 10 most common perceived **stereotypes for the social group "Politician"** among the population of Michigan State Undergraduates are `corrupt, liars, greedy, rude, rich, selfish, old, smart, white,` and `moneyhungry`.

The 10 most common perceived **counter-stereotypes for the social group "Politician"** among the population of Michigan State Undergraduates are `honest, caring, kind, empathy, honesty, quiet, shy, poor, kindness, ` and `selfless`.

The 10 most common **non-stereotypic traits** were `hungry, thirsty, tired, happy, sad, athletic, sleepy, tall, angry,` and `funny`.

Interestingly, at first glance, it seems that stereotypes and counterstereotypes have a very high degree of antonymy, so further (more formal) analysis can be done comparing the frequency of stereotypes and countersterotypes and their degree of antonymy.
